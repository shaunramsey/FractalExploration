```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
def bifurcation_plot_eb(x_0, savepng = False):
    
    r = np.arange(3.4, 3.6, .0001)
    
    plt.figure(figsize = (11,8.5))
    
    n = 100
    bf_x = [[] for _ in range(n)]
    
    for i in range(len(r)):
        x = x_0
        
        for j in range(700):
            x = r[i] * x * (1 - x)
        
        for j in range(n):
            bf_x[j].append(x)
            x = r[i] * x * (1 - x)  
            
        
    
    # plots a line that goes though the arrays of points constructed
    for i in range(n):
        plt.plot(r, bf_x[i], c = 'black', lw = .5)
    
    # set y and x ticks
   # plt.xticks(np.linspace(2,4,11))
    plt.yticks(np.linspace(0,1,11)) 
    plt.grid()

    plt.xlabel('r')
    plt.ylabel('x')
    
    plt.title("$x_0$ =" + str(x_0))
    
    plt.show()
    
    if(savepng):
        plt.savefig("bifurcation_diagram_eb.png", format="png")
    
```

```python
bifurcation_plot_eb(.25, True)
```

```python
def bifurcation_plot(x_0, savepng = False):
    
    r = np.arange(3.4, 3.6, .0001)
    
    plt.figure(figsize = (11,8.5))
    
    x = x_0
    
    for i in range(200):
        x = np.multiply(x, r) * np.subtract(1,x)
    
    for i in range(100):
        x = np.multiply(x, r) * np.subtract(1,x)
        plt.scatter(r,x,s =1)
        
    
    # set y and x ticks
    # plt.xticks(np.linspace(1,4,11))
    plt.yticks(np.linspace(0,1,11)) 
    plt.grid()

    plt.xlabel('r')
    plt.ylabel('x')
    
    plt.title("$x_0$ =" + str(x_0))
    
    plt.show()
    
    if(savepng):
        plt.savefig("bifurcation_diagram.png", format="png")
```

```python
bifurcation_plot(.25, True)
```
